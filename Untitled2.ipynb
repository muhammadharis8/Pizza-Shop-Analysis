```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['day_of_week'] = df['order_date'].dt.day_name()


busiest_days = df.groupby('day_of_week')['order_id'].nunique().sort_values(ascending=False)


best_selling_pizzas = df.groupby('pizza_name')['quantity'].sum().sort_values(ascending=False)

# Extract the hour from the order_time and calculate total pizzas made during peak hours
df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M:%S').dt.hour
peak_hours = df.groupby('order_time')['quantity'].sum().sort_values(ascending=False)

busiest_days, best_selling_pizzas.head(), peak_hours.head()
```
